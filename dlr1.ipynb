```python
import numpy as np
from numpy import linalg as la
import matplotlib.pyplot as plt
import math
import pylab
from mpl_toolkits.mplot3d import Axes3D
```

```python
 fa = lambda a: a
fu0 = lambda h, i: 1
u_x0 = lambda x: 0
```

```python
def sch(t, h, left, right, k, a):
    tsize = 1 + int(1 / t)  
    hsize = k 
    x = np.linspace(left, right, k)
    u = np.zeros((tsize, hsize))
    u0 = np.array([u_x0(i) for i in np.linspace(left, right, k)])
    l=r=0
    for i in range(0, k-1):
        if(x[i] == 0.0):
            l=i
        if(x[i] == 1.0):
            r=i          
    for i in range(l, r): 
        u[0][i] = fu0(h, i)  
    for i in range(tsize-1):
        for j in range(1,hsize-1):
            if(a>=0):
                u[i+1, j]=u[i, j]-a*t*(u[i, j]-u[i, j-1])/h
            elif(a<0):
                u[i+1, j]=u[i, j]-a*t*(u[i, j+1]-u[i, j])/h
    return u
```

```python
h = 0.05
t = 0.05
left = -2
right = 2
k = int((right-left)/h) + 1
a = -1

u_sch = sch(t, h, left, right, k, a)
plt.plot(u_sch[0])
```
